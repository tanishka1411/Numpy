Format Strings

A format string consists of a part for color, marker and line:

`fmt = '[marker][line][color]'`

|character|description|
|:-------|:--------|
|'.'|	point marker|
|','|	pixel marker|
|'o'|	circle marker|
|'v'|	triangle_down marker|
|'^'|	triangle_up marker|
|'<'|	triangle_left marker|
|'>'|	triangle_right marker|
|'1'|	tri_down marker|
|'2'|	tri_up marker|
|'3'|	tri_left marker|
|'4'|	tri_right marker|

**Line Styles**

|character|	description|
|:---------|:-----------|
|'-'|	solid line style|
|'--'|	dashed line style|
|'-.'|	dash-dot line style|
|':'|	dotted line style|

link for formatting styles:
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
